# **Association Rule Mining | Apriori Algorithm** 

##  **Association Rule Learning**:

* Assocation Rule Learning is a type of unsupervised learning technique.
* It Checks the dependency of one data item on another data item and maps accordingly so that it can be more profitable.
* It tries to find some interesting relations or associations among the varables of dataset.
  

### **Example**: Market Basket Analysis


* In supermarket,all products that are purchased together are put together.
* If a customer buys bread, he most likely can also buy butter, eggs or milk so these products are stored within a shelf or mostly nearby.

### **How does Assocation Rule Learning Work ?**

* Association Rule learning works on the concept of IF and ELSE Statement ,such as if A then B.
  

* Here the **IF** element is callled **antecedent** , and **then** statement is called as **Cosequent**.
* These types of relationships where we can find out some association or relation between two items is known as **single cardinatlity**

### **Association Rule Evaluation**

## 1. Support :

* Support is the frequency of A  or **how frequently an item apperas** in the dataset.
* It is defined as the fraction of the transaction T that contains the items X.
  
  

### 2. Confidence :

* Confidence indicates how to often the rule has been found to be true.
* or how often the items **X and Y occur together** in the dataset when the occurence of X is already given.
